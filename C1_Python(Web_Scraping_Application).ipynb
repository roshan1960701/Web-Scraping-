# Web Scraping
Request library has been used in web scraping to send HTTP requests and get back the response in form of the HTML tags. If the response code is 200 it indicates that the request has succeeded. For extracting data from the HTML tags, the Beautiful Soup package is being implemented. The find and find_all functions allow us to extract data from particular tags

```python
url='https://www.ecrater.com/c/76733/antiques?&perpage=80'
resp=requests.get(url)
```

```python
resp
```

```python
resp.content
```

```python
soup=BeautifulSoup(resp.content)
```

```python
print(soup.prettify())
```

```python
ul=soup.find('ul',{'id':"product-list-grid",})
```

```python
print(ul)
```

```python
listItems=ul.find_all('li',{'class':"product-item"})
```

```python
len(listItems)
```

# Data Cleaning
In this process, Data is being extracted from HTML tags, although, price has dollar($) sign which has to be replaced with empty space to perform mathematical calculations. The get function has been consumed to obtain image source from Image tag

```python
productList=[]
for item in listItems:
  record={}
  record['ProductName']=item.find('div',{'class':"product-details"}).find('h2').text
  record['Price (Dollar)']=float(item.find('span',{'class':"btn btn-inverse"}).text.replace('$','').replace(',',''))
  record['productImage'] = item.find('a',{'class':"product-image"}).find('img').get('src')
  record['Category'] = item.find('div', attrs={'class' : 'product-details'}).find('a', attrs={'class':'product-category'}).text
  # category = item.find('span',{'class':"product-category"})
  # if category is not None:
  #   record['Category'] = item.find('div',{'class':"product-category"}).text
  # else:
  #   record['Category']= None

  #category = item.find('div', attrs={'class' : 'product-details'}).find('a',{'class':'product-category'})

  # used to get link from product
  # category = item.find('div', attrs={'class' : 'product-details'}).find('a',{'class':'product-category'}).get("href")

  # category1 =  item.find('a', attrs={'class':'product-category'}).text

# soup.find("a",{"class":" class of your target a element"}).get("href")
  productList.append(record)
```

The json normalizing simply convert raw data into dataset format so, that we could perform mathematical calculations on it. To display basic mathematical operations such as count, mean, sd, min and max, decribe() function has been consumed.

### Displaying Data in HTML Table format
The Normal Dataset format does not support images to overcome this problem to_html() method has been implemented. Ipython.core.display package is used to consume HTML() for displaying images with its other features. The Html function needs images in Img tag with Image source. For converting image link to html Img tag with source, I have created  path_to_image_html which will return as per the specific format to display image in table.

```python
df = pd.DataFrame.from_records(productList)
```

```python
def path_to_image_html(path):
    return '<img src="'+ path + '" width="60" >'
```

```python
df.to_html(escape=False, formatters=dict(productImage=path_to_image_html))
```

```python
from IPython.core.display import HTML
HTML(df.to_html(escape=False,formatters=dict(productImage=path_to_image_html)))
```

### Find products which have price less than 100 and greater than 10 Dollar  

```python
q1= ecrater.loc[(ecrater['Price (Dollar)'] > 10) & (ecrater['Price (Dollar)'] < 100)]
df = pd.DataFrame.from_records(q1)
df.to_html(escape=False, formatters=dict(productImage=path_to_image_html))
HTML(df.to_html(escape=False,formatters=dict(productImage=path_to_image_html)))
```
